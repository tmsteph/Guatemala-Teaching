```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
def scrape_category():
    books = []
    for x in range(10):
        # HTML object
        html = browser.html
        # Parse HTML with Beautiful Soup
        soup = BeautifulSoup(html, 'lxml')
        # Retrieve all elements that contain book information
        articles = soup.find_all('article', class_='product_pod')

        # Iterate through each book
        for article in articles:
            # Use Beautiful Soup's find() method to navigate and retrieve attributes
            h3 = article.find('h3')
            link = h3.find('a')
            
            book = {}
            book['url'] = ('http://books.toscrape.com/' + link['href'])
            book['title'] = link['title']
            print('-----------')
            print(book)
            books.append(book)

        # Click the 'Next' button on each page
        try:
            browser.links.find_by_partial_text('next').click()

        except Exception: # ElementDoesNotExist:
            print("Scraping Complete")
            break
    return books
```

```python
html = browser.html
soup = BeautifulSoup(html, 'lxml')

sidebar = soup.find('ul', class_='nav-list')

category_list = sidebar.ul.find_all('li')

categories = []
for category in category_list:
    category_name = category.a.text.strip()
    url = 'http://books.toscrape.com/' + category.a['href']
    categories.append( (category_name, url))

results = {}
try:
    for (category, url) in categories:
        print(f"Scraping category {category}")
        print(f"Visiting url {url}")
        browser.visit(url)
        books = scrape_category()
        results[category] = books
except Exception as e:  # ElementDoesNotExist:
    print("Problem listing categories")
    raise e
```

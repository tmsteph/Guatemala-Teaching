```python
# Dependencies
import requests
import json
```

```python
# URL for GET requests to retrieve vehicle data
url = "https://api.spacexdata.com/v3/launchpads"
```

```python
# Pretty print JSON for all launchpads
raw_response = requests.get(url)
print(type(raw_response))
print(help(raw_response))

```

```python
response = raw_response.json()
print(type(response))
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# Pretty print JSON for a specific launchpad
response2 = requests.get(url + "/ksc_lc_39a").json()
print(json.dumps(response2, indent=4, sort_keys=True))
```

```python
# Alternately, pull the launchpad info from the original request
print(json.dumps(response[5], indent=4, sort_keys=True))
```
